```
import pandas as pd, numpy as np, matplotlib.pyplot as plt
df = pd.read_csv('doctor_train.csv')

Xdf = df.drop('Y',axis=1)
Ydf = df['Y']

test_df = pd.read_csv('doctor_test.csv')
```

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(Ydf)
Y = le.transform(Ydf)

num_cal = ['age','Money','day','Time','Doctor_visits','last_visit','cured_in']
cat_col = ['Profession','Status','edu','Irregular','residence','prev_diagnosed','communication','Month','side_effects']

Xdfm = Xdf.copy()
Xdfm = Xdfm.fillna(0)
Xodfm = Xdfm.copy()
Xodfm = pd.get_dummies(Xodfm)

tdfm = test_df.copy()
tdfm = tdfm.fillna(0)
todf = tdfm.copy()
todf = pd.get_dummies(todf)
```

```
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

X_train, X_test, y_train, y_test = train_test_split(Xodfm, Y, test_size=0.1, random_state=0, stratify=Ydf)
```

```
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```
rfc = RandomForestClassifier()
rfc.fit(Xodfm,Y)
y_pred = rfc.predict(todf)

le_y = le.inverse_transform(y_pred)

output = pd.DataFrame({'id': todf.ID,
                       'Y': le_y})
output.to_csv('submission.csv', index=False)
from google.colab import files
files.download('submission.csv')
```
