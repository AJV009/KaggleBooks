# feature engineering

```

import featuretools as ft
import numpy as np
import pandas as pd
df = pd.read_csv('doctor_train.csv')
# creating and entity set 'es'
es = ft.EntitySet(id = 'doctor')
# adding a dataframe 
es = es.entity_from_dataframe(entity_id = 'doctor', dataframe = df, index = 'ID')
```

```
primitives = ft.list_primitives()
pd.options.display.max_colwidth = 100
agg_prim = primitives[primitives['type'] == 'aggregation']['name'].values
trans_prim = primitives[primitives['type'] == 'transform']['name'].values
```

```
features, feature_names = ft.dfs(entityset = es, target_entity = 'doctor', max_depth=5, max_features=300, ignore_entities=['Y'],
                                 agg_primitives=agg_prim, trans_primitives=trans_prim)
```
