```
path = Path('/content')
path.ls()
```

```
df = pd.read_csv(path/'menergy_train.csv')
df.head()
```

```
# dls = TabularDataLoaders.from_csv(path/'adult.csv', path=path, y_names="salary",
#     cat_names = ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race'],
#     cont_names = ['age', 'fnlwgt', 'education-num'],
#     procs = [Categorify, FillMissing, Normalize])

dls = TabularDataLoaders.from_csv(path/'menergy_train.csv', path=path, y_names="WattHour",
    # cat_names = ['luminousity'],
    cont_names = ['luminousity','degree_C1','moisture_1','degree_C2','moisture_2','degree_C3','moisture_3','degree_C4','moisture_4','degree_C5','moisture_5','degree_C6','moisture_6','degree_C7','moisture_7','degree_C8','moisture_8','degree_C9','moisture_9','degree_Cout','Pressure','moisture_out','Wind','Clarity','dew_index'],
    # procs = [Normalize]
    )
```

```
dls.show_batch()
```

```
learn = tabular_learner(dls, metrics=accuracy, loss_func=F.mse_loss)
```

```
learn.fit_one_cycle(1)
```

```
learn.predict(df.iloc[0])
```

```
test_df = df.copy()
test_df.drop(['salary'], axis=1, inplace=True)
dl = learn.dls.test_dl(test_df)
```

```
learn.get_preds(dl=dl)
```
