```
import pandas as pd
df = pd.read_csv('energy_train.csv')
test_df = pd.read_csv('energy_test.csv')

df = df.fillna(df.mean())
test_df = test_df.fillna(test_df.mean())

y = df['WattHour']
X = df.drop(['WattHour','id','random_variable_1', 'random_variable_2'],axis=1)
```

```
cpd = df.copy()

cpd['mdegree'] = cpd[['degree_C1','degree_C2','degree_C3','degree_C4','degree_C5',
                           'degree_C6','degree_C7','degree_C8','degree_C9']].mean(axis=1)

cpd['mmoisture'] = cpd[['moisture_1','moisture_2','moisture_3','moisture_4','moisture_5','moisture_6','moisture_7','moisture_8','moisture_9']].mean(axis=1)

dm = ['degree_C1','degree_C2','degree_C3','degree_C4','degree_C5',
                           'degree_C6','degree_C7','degree_C8','degree_C9',
      'moisture_1','moisture_2','moisture_3','moisture_4','moisture_5','moisture_6',
      'moisture_7','moisture_8','moisture_9']
```

```
pdf = df.copy()
pdf["date"] = pd.to_datetime(pdf["date"])
pdf["month"] = pdf["date"].dt.month
pdf["day"] = pdf["date"].dt.day
pdf["hour"] = pdf["date"].dt.hour
pdf["minute"] = pdf["date"].dt.minute
emdf = pdf
for i in range(1,10):
    emdf['EMA_C'+str(i)] = emdf['degree_C'+str(i)].ewm(span=10,adjust=False).mean()
    emdf['EMA_M'+str(i)] = emdf['moisture_'+str(i)].ewm(span=10,adjust=False).mean()
emdf['EMA_degree_Cout'] = emdf["degree_Cout"].ewm(span=10,adjust=False).mean()
emdf['EMA_Pressure'] = emdf["Pressure"].ewm(span=10,adjust=False).mean()
emdf['EMA_moisture_out'] = emdf["moisture_out"].ewm(span=10,adjust=False).mean()
emdf['EMA_Wind'] = emdf["Wind"].ewm(span=10,adjust=False).mean()
emdf['EMA_Clarity'] = emdf["Clarity"].ewm(span=10,adjust=False).mean()
emdf['EMA_dew_index'] = emdf["dew_index"].ewm(span=10,adjust=False).mean()

cols_x = ['hour','WattHour','luminousity','degree_C2','EMA_C2','EMA_C6','degree_C6','degree_C3',
             'degree_Cout','EMA_degree_Cout','EMA_C3','moisture_1','EMA_Wind','Wind',
             'degree_C1','degree_C4','degree_C8','EMA_M1','EMA_C1','degree_C7','EMA_C4',
             'EMA_C8','EMA_C7','moisture_3','degree_C5','EMA_C9','degree_C9','EMA_M3',
             'EMA_dew_index','dew_index','EMA_C5','minute','month']

hcols_x = ['hour','WattHour','EMA_C2','EMA_C6','luminousity','degree_C2','degree_C6','degree_C3','degree_Cout']

odf = emdf
cdf = emdf[cols_x]
hdf = emdf[hcols_x]

y = odf['WattHour'].copy()
X = odf.drop(columns=['WattHour','date','id']).copy()

pdf = test_df.copy()
pdf["date"] = pd.to_datetime(pdf["date"])
pdf["month"] = pdf["date"].dt.month
pdf["day"] = pdf["date"].dt.day
pdf["hour"] = pdf["date"].dt.hour
pdf["minute"] = pdf["date"].dt.minute
emdf = pdf
for i in range(1,10):
    emdf['EMA_C'+str(i)] = emdf['degree_C'+str(i)].ewm(span=10,adjust=False).mean()
    emdf['EMA_M'+str(i)] = emdf['moisture_'+str(i)].ewm(span=10,adjust=False).mean()

emdf['EMA_degree_Cout'] = emdf["degree_Cout"].ewm(span=10,adjust=False).mean()
emdf['EMA_Pressure'] = emdf["Pressure"].ewm(span=10,adjust=False).mean()
emdf['EMA_moisture_out'] = emdf["moisture_out"].ewm(span=10,adjust=False).mean()
emdf['EMA_Wind'] = emdf["Wind"].ewm(span=10,adjust=False).mean()
emdf['EMA_Clarity'] = emdf["Clarity"].ewm(span=10,adjust=False).mean()
emdf['EMA_dew_index'] = emdf["dew_index"].ewm(span=10,adjust=False).mean()

cols_x = ['hour','luminousity','degree_C2','EMA_C2','EMA_C6','degree_C6','degree_C3',
             'degree_Cout','EMA_degree_Cout','EMA_C3','moisture_1','EMA_Wind','Wind',
             'degree_C1','degree_C4','degree_C8','EMA_M1','EMA_C1','degree_C7','EMA_C4',
             'EMA_C8','EMA_C7','moisture_3','degree_C5','EMA_C9','degree_C9','EMA_M3',
             'EMA_dew_index','dew_index','EMA_C5','minute','month']

hcols_x = ['hour','EMA_C2','EMA_C6','luminousity','degree_C2','degree_C6','degree_C3','degree_Cout']

odf = emdf.drop(columns=['date','id'])
cdf = emdf[cols_x]
hdf = emdf[hcols_x]
```
